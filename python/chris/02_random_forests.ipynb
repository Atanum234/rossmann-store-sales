```python
%matplotlib inline

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import matplotlib.cm as cm

train = pd.read_csv('../data/train.csv', parse_dates = ['Date']).sort_values(by='Date')
store = pd.read_csv('../data/store.csv')
test = pd.read_csv('../data/test.csv', parse_dates = ['Date']).sort_values(by='Date')
```

```python
train.head()
```

```python
train['year'] = train.Date.apply(lambda x: x.year)
train['month'] = train.Date.apply(lambda x: x.month)
train['woy'] = train.Date.apply(lambda x: x.weekofyear)
train.drop(['Date'], axis = 1, inplace= True)
train.head()
```

```python
train = train.merge(store, on='Store', copy=False)
train.head()
```

```python
test = test.merge(store, on='Store', copy=False)
test.head()
```

```python
train['CompetitionOpen'] = 12 * (train.year - train.CompetitionOpenSinceYear) + (train.month - train.CompetitionOpenSinceMonth)
train['CompetitionOpen'] = train.CompetitionOpen.apply(lambda x: x if x > 0 else 0)
train.drop(['CompetitionOpenSinceMonth', 'CompetitionOpenSinceYear'], axis = 1, inplace = True)
train.head()
```

```python
train['PromoOpen'] = 12 * (train.year - train.Promo2SinceYear) + (train.woy - train.Promo2SinceWeek) / float(4)
train['PromoOpen'] = train.CompetitionOpen.apply(lambda x: x if x > 0 else 0)
train.drop(['Promo2SinceYear', 'Promo2SinceWeek'], axis = 1, inplace = True)
train.head()
```

```python
train = train.fillna(0)
train = train.sort_index(axis=1)
```

```python
train = pd.get_dummies(train, columns = ['StateHoliday', 'StoreType', 'Assortment'])
train.drop(['StateHoliday', 'StoreType', 'Assortment', 'PromoInterval'], axis = 1, inplace = True)
train.head()
```
