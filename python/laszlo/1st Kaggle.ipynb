```python
import pandas as pd
train = pd.read_csv('data/train.csv')
test = pd.read_csv('data/test.csv')
store = pd.read_csv('data/store.csv')
```

```python
train.describe()
```

```python
train.head()
```

```python
store[0:3]
```

```python
train = train.merge(store, on='Store', copy=False)
train.head()
```

```python
train = train.fillna(0)
train.head()
```

```python
train.drop(['Date'], axis=1, inplace = True)
```

```python
train['StateHoliday'] = train['StateHoliday'].apply(lambda x: 0 if '0' == x else 1)
train.head()
```

```python
from sklearn.ensemble import RandomForestRegressor
rfr = RandomForestRegressor(n_estimators=30)
```

```python
X_train = train.drop(['Sales'], axis=1).values
y_train = train['Sales'].values
```

```python
X_train
```

```python
train_cat=pd.get_dummies(train, columns=['StoreType'])
```

```python
train_cat.head()
```

```python
from sklearn import cross_validation as cv
scores = cv.cross_val_score(rfr, X_train, y_train, scoring='mean_squared_error', cv=4)
```
